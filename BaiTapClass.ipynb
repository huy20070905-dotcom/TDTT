```python
class Cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def s_be_mat(self):
        return 2*3.14*self.h+2*3.14*self.r**2
    def the_tich(self):
        return 3.14*self.r**2*self.h
r,h=map(int,input().split())
giatri=Cylinder(r,h)
stp=giatri.s_be_mat()
thetich=giatri.the_tich()
print(f"{stp:.2f}")
print(f"{thetich:.2f}")
```

```python
class Date:
    def __init__(self,day,month,year):
        self.day=day
        self.month=month
        self.year=year
    def nam_nhuan(self):
        return (self.year%4==0 and self.year%100!=0)or(self.year%400==0)
    def kiem_tra(self):
        if self.day<1 or self.year<1 or self.month<1 or self.month>12:
            return False
        thang = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.nam_nhuan():
            thang[2]=29
        return self.day<=thang[self.month]
    def ngay_ke_tiep(self):
        thang = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.nam_nhuan():
            thang[2]=29
        next_day=self.day+1
        next_month=self.month
        next_year=self.year
        if next_day>thang[self.month]:
            next_day=1
            next_month+=1
            if next_month>12:
                next_year+=1
        return Date(next_day,next_month,next_year)
    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year:04d}"
du_lieu=input().split('/')
if len(du_lieu)==3 and du_lieu[0].isdigit() and du_lieu[1].isdigit() and du_lieu[2].isdigit():
    d=int(du_lieu[0])
    m=int(du_lieu[1])
    y=int(du_lieu[2])
    my_date=Date(d,m,y)
    if my_date.kiem_tra():
        print(my_date.ngay_ke_tiep())
    else:
        print("INVALID")
else:
    print("INVALID")
```

```python
import math
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def goc_toa_do(self):
        return self.x==0 and self.y==0
    def tren_truc_tung(self):
        return self.x==0
    def tren_truc_hoanh(self):
        return self.y==0
    def khoang_cach(self,orther_point):
        return math.sqrt((orther_point.x-self.x)**2+(orther_point.y-self.y)**2)
x,y=map(float,input().split())
M=Point(x,y)
origin=Point(0,0)
if M.goc_toa_do():
    print("M là gốc tọa độ")
else:
    if M.tren_truc_tung():
        print("M nằm trên trục tung")
    if M.tren_truc_hoanh():
        print("M nằm trên trục hoành")
    if not M.tren_truc_hoanh() and not M.tren_truc_tung():
        print("M không nằm trên trục tọa độ")
dist=M.khoang_cach(origin)
print("Khoảng cách từ M đến gốc tọa độ là: "f"{dist:.2f}")
```

```python
class Calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def subtract(self):
        return self.a-self.b
    def product(self):
        return self.a*self.b
    def divide(self):
        if self.b ==0:
            return "Lỗi: chia cho 0"
        return self.a/self.b
    def power(self):
        return self.a**self.b
    def mod(self):
        if self.b==0:
            return "Lỗi: chia cho 0"
        return self.a%self.b
    def set_numbers(self,a,b):
        self.a=a
        self.b=b
calc=Calculator(0,0)
print("---MÁY TÍNH BỎ TÚI---")    
print("Chọn phép tính: 1.Cộng, 2.Trừ, 3.Nhân, 4.Chia, 5.Lũy thừa, 6.Chia dư")
while True:
    a,b=map(float,input("Nhập 2 số cần tính: ").split())
    calc.set_numbers(a,b)
    choice=input()
    if choice=='1':
        print(calc.add())
    elif choice=='2':
        print(calc.subtract())
    elif choice=='3':
        print(calc.product())
    elif choice=='4':
        print(calc.divide())
    elif choice=='5':
        print(calc.power())
    elif choice=='6':
        print(calc.mod())
    else:
        print("Lựa chọn không hợp lệ")
    out=input("Bạn có muốn thoát không? YES để dừng: ")
    if out=='YES':
        break
```

```python
class ShoppingCart:
    def __init__(self):
        self.items=[]
    def add_item(self,name,price):
        self.items.append({"name":name,"price":price})
        print("Đã thêm")
    def remove(self,name):
        for i in self.items:
            if i["name"]==name:
                self.items.remove(i)
                print("Đã xóa")
                return
            else:
                print("Không tìm thấy sản phẩm để xóa")
    def check(self):
        return len(self.items)==0
    def total_price(self):
        cnt=0
        for i in self.items:
            cnt+=i["price"]
        return cnt
    def kiem_tra(self):
        if self.check():
            print("Giỏ hàng trống")
        else:
            i=1
            for item in self.items:
                print(f"{i}. {item['name']}: {item['price']}")
                i+=1
    def clear(self):
        self.items=[]
        print("Đã xóa danh sách")
cart=ShoppingCart()
print("---QUẢN LÝ GIỎ HÀNG---")
print("1. Thêm sản phẩm\n2. Xóa sản phẩm\n3. Kiểm tra giỏ hàng rỗng\n4. Tổng tiền\n5. Hiển thị danh sách\n6. Xóa toàn bộ giỏ hàng")
while True:
    choice=input("Nhập lựa chọn: ")
    if choice=='1':
        name=input("Tên: ")
        price=float(input("Giá: "))
        cart.add_item(name,price)
    elif choice=='2':
        name=input("Tên cần xóa: ")
        cart.remove(name)
    elif choice=='3':
        if cart.check():
            print("Giỏ hàng rỗng")
        else:
           print("Giỏ có sản phẩm")
    elif choice=='4':
        print(f"Tổng tiền: {cart.total_price()}")
    elif choice=='5':
        cart.kiem_tra()
    elif choice=='6':
        cart.clear()
    else:
        print("Không hợp lệ")
    out= input("\nBạn có muốn thoát chương trình không? (YES để thoát,NO để tiếp tục): ")
    if out== 'YES':
        break
    elif out=='NO':
        continue
```

```python
class Rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,k):
        self.w*=k
        self.h*=k
w,h,k=map(float,input().split())
kq=Rectangle(w,h)
kq.scale(k)
print(kq.area(),kq.perimeter())
```

```python
import math
class Fraction:
    def __init__(self,num,den):
        commom=math.gcd(num,den)
        self.num=num//commom
        self.den=den//commom
        if self.den<0:
            self.num=-self.num
            self.den=-self.den
    def add(self,orther):
        return Fraction(self.num*orther.den+orther.num*self.den,self.den*orther.den)
    def sub(self,orther):
        return Fraction(self.num*orther.den-orther.num*self.den,self.den*orther.den)
    def mul(self,orther):
        return Fraction(self.num*orther.num,self.den*orther.den)
    def div(self,other):
        return Fraction(self.num*orther.den,self.den*orther.num)
    def __str__(self):
        return f"{self.num}/{self.den}"
du_lieu=input().split()
a,b,op,c,d=int(du_lieu[0]),int(du_lieu[1]),du_lieu[2],int(du_lieu[3]),int(du_lieu[4])
ps1=Fraction(a,b)
ps2=Fraction(c,d)
if op=='+':
    print(ps1.add(ps2))
elif op=='-':
    print(ps1.sub(ps2))
elif op=='*':
    print(ps1.mul(ps2))
elif op=='/':
    print(ps1.div(ps2))
```

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def deposit(self,x):
        self.balance+=x
    def withdraw(self,x):
        if x<=self.balance:
            self.balance-=x
    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"
du_lieu=input().split()
owner=du_lieu[0]
initial_blance=float(du_lieu[1])
acount=BankAccount(owner,initial_blance)
n=int(input())
for i in range(n):
    cmd=input().split()
    action=cmd[0]
    amount=float(cmd[1])
    if action=="DEPOSIT":
        acount.deposit(amount)
    elif action=="WITHDRAW":
        acount.withdraw(amount)
print(f"{acount.balance:.2f}")
```

```python
class Student:
    def __init__(self,name):
        self.name=name
        self.score=[]
    def add_score(self,subject,score):
        self.score.append(score)
    def avg(self):
        if len(self.score)==0:
            return 0.0
        return sum(self.score)/len(self.score)
    def rank(self):
        average=self.avg()
        if average >= 8.0:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5.0:
            return "Average"
        else:
            return "Poor"
name=input()
n=int(input())
student=Student(name)
for i in range(n):
    line=input().strip().split()
    score=float(line[-1])
    subject=" ".join(line[:-1])
    student.add_score(subject,score)
print(f"{student.name} {student.avg():.2f} {student.rank()}")
```

```python
class Book:
    def __init__(self,title,author,year):
        self.title=title
        self.author=author
        self.year=int(year)
class Library:
    def __init__(self):
        self.books=[]
    def add(self,book):
        self.books.append(book)
    def count_by_author(self,name):
        cnt=0
        for book in self.books:
            if book.author==name:
                cnt+=1
        return cnt
    def find_by_year(self,y):
        cnt=0
        for book in self.books:
            if book.year==int(y):
                cnt+=1
        return cnt
n=int(input().strip())
library=Library()
for i in range(n):
    parts=input().strip().split(maxsplit=1)
    cmd=parts[0]
    if cmd=="ADD":
        data=parts[1].split(';')
        title=data[0]
        author=data[1]
        year=int(data[2])
        library.add(Book(title,author,year))
    elif cmd == "COUNT":
        author_name = parts[1]
        print(library.count_by_author(author_name))   
    elif cmd == "COUNTYEAR":
        year = int(parts[1])
        print(library.find_by_year(year))
```

```python
import math
class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def __sub__(self, other):
        return Vector3D(self.x - other.x, self.y - other.y, self.z - other.z)
    def dot(self, other):
        return (self.x * other.x) + (self.y * other.y) + (self.z * other.z)
    def cross(self, other):
        return Vector3D(self.y * other.z - self.z * other.y, self.z * other.x - self.x * other.z, self.x * other.y - self.y * other.x)
    def absolute(self):
        return math.sqrt(self.x ** 2 + self.y ** 2 + self.z ** 2)
points = []
for _ in range(4):
    x, y, z = map(float, input().strip().split())
    points.append(Vector3D(x, y, z))
A, B, C, D = points
AB = B - A
BC = C - B
CD = D - C
X = AB.cross(BC)
Y = BC.cross(CD) 
dot_product = X.dot(Y)
mag_product = X.absolute() * Y.absolute()
cos_phi = max(-1.0, min(1.0, dot_product / mag_product))
phi_deg = math.degrees(math.acos(cos_phi))
print(f"{phi_deg:.2f}")
```

```python
class ComplexNumber:
    def __init__(self,initRealPart,initImaginaryPart):
        self.realPart=float(initRealPart)
        self.imaginaryPart=float(initImaginaryPart)
    def multiply(self,other):
        real = (self.realPart * other.realPart) - (self.imaginaryPart * other.imaginaryPart)
        imag = (self.realPart * other.imaginaryPart) + (self.imaginaryPart * other.realPart)
        return ComplexNumber(real, imag)
r1, i1 = map(float, input().strip().split())
r2, i2 = map(float, input().strip().split())
c1 = ComplexNumber(r1, i1)
c2 = ComplexNumber(r2, i2)
result = c1.multiply(c2)
print(f"{int(result.realPart)} {int(result.imaginaryPart)}")
```

```python
import math
class Triangle:
    def __init__(self,egde1,egde2,egde3):
        self.egde1=egde1
        self.egde2=egde2
        self.egde3=egde3
    def getArea(self):
        if (self.egde1+self.egde2>self.egde3) and (self.egde3+self.egde2>self.egde2) and (self.egde1+self.egde3>self.egde3):
            p=(self.egde1+self.egde2+self.egde3)/2
            area=math.sqrt(p*(p-self.egde1)*(p-self.egde2)*(p-self.egde3))
            return area
        else:
            return -1
du_lieu=input().split()
e1=du_lieu[0]
e2=du_lieu[1]
e3=du_lieu[2]
tam_giac=Triangle(e1,e2,e3)
area=tam_giac.getArea()
if area==-1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
