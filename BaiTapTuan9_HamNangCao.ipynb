```python
def merge_recursive(L1,L2):
    if not L1: return L2
    if not L2: return L1
    if L1[0]<=L2[0]:
        return [L1[0]]+merge_recursive(L1[1:],L2)
    else:
        return [L2[0]]+merge_recursive(L1,L2[1:])
L1=list(map(int,input().split()))
L2=list(map(int,input().split()))
print(merge_recursive(L1,L2))
```

```python
def minimize_by_one_swap(s,idx=0):
    if idx>=len(s)-1:
        return "".join(s)
    min_val=s[idx]
    min_pos=idx
    for i in range(len(s)-1,idx,-1):
        if s[i]<min_val and not (idx==0 and s[i]=='0'):
            min_val=s[i]
            min_pos=i 
    if min_pos!=idx:
        s[idx],s[min_pos]=s[min_pos],s[idx]
        return "".join(s)
    return minimize_by_one_swap(s,idx+1)
s=list(input())
print(minimize_by_one_swap(s))
```
