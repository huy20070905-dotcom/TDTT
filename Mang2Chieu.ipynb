Bài 2A: Tính tổng cột k của ma trận nxm số nguyên (n, m vào từ bàn phím)

```python
n=int(input())
m=int(input())
matrix=[]
for i in range(n):
    row=[]
    for j in range(m):
        val=int(input())
        row.append(val)
    matrix.append(row)
k=int(input())
if 1<=k<=m:
    total=0
    col_index=k-1
    for i in range(n):
        total+=matrix[i][col_index]
    print(total)
```

Bài 2B: Viết hàm zigzag_numbers(n) trả về một chuỗi biểu diễn ma trận n x n chứa các số từ 1 đến n^2. Các hàng cách nhau bởi ký tự xuống 
dòng. Với chỉ số hàng bắt đầu từ 0: hàng chẵn đi từ trái sang phải, hàng lẻ đi từ phải sang trái. Các số trong cùng một hàng cách nhau bởi 
một dấu cách. 
n = 2
1 2
4 3
n= 4
1 2 3 4
8 7 6 5
9 10 11 12
16 15 14 13

```python
def zigzag_numbers(n):
    matrix=[]
    cnt=1
    for i in range(n):
        row=[]
        for j in range(n):
            row.append(cnt)
            cnt+=1
        matrix.append(row)
    kq=[]
    for i in range(n):
        if i%2==0:
            row_data=matrix[i]
        else:
            row_data=reversed(matrix[i])
        kq.append(" ".join(map(str,row_data)))
    return "\n".join(kq)
n=int(input())
print(zigzag_numbers(n))
```

Bài 3: Trả về cột (chỉ số bắt đầu từ 0) có tổng cột lớn nhất trong ma trận n x m. 
Viết hàm column_with_max_sum(mat) trả về chỉ số bắt đầu từ 0 của cột có tổng lớn nhất. 
Nếu có nhiều cột cùng tổng lớn nhất, trả về cột có chỉ số nhỏ nhất. - Dùng vòng lặp để tính tổng từng cột. - Không dùng tuple hoặc dictionary. 
n = 2
m = 3
mat = [[1, 3, 5], 
       [6, 2, -9]]


```python
n=int(input())
m=int(input())
mat=[]
for i in range(n):
    row=[]
    for j in range(m):
        val=int(input())
        row.append(val)
    mat.append(row)
def column_with_max_sum(mat):
    kq=[]
    for i in range(m):
        cnt=0
        for j in range(n):
            cnt+=mat[j][i]
        kq.append(cnt)
    index=kq.index(max(kq))
    return index
print(column_with_max_sum(mat))
```

Bài 6: Cho ma trận nxm số nguyên, thêm vào một cột tại vị trí k.

```python
n=int(input())
m=int(input())
matrix=[]
for i in range(n):
    row=[]
    for j in range(m):
        val=int(input())
        row.append(val)
    matrix.append(row)
k=int(input())
new_col=[]
for i in range(n):
    index=int(input())
    new_col.append(index)
for i in range(n):
    matrix[i].insert(k,new_col[i])
print(matrix)
```
