# W10A10

```python
try:
    N=int(input())
    arr=list(map(int,input("Nhập vào N phần tử").split()))
    if N<=0 or len(arr)!=N:
        raise ValueError
    else:
        for i in range(len(arr)):
            for j in range(i+1,len(arr)):
                if arr[i]==arr[j]:
                    raise ValueError
        print(sum(arr))
except ValueError:
    print("Mang khong hop le")
```
