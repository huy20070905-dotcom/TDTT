# W7A19

```python
arr1=list(map(int,input().split()))
arr2=list(map(int,input().split()))
def merge(arr1,arr2):
    def chuanhoa(mang):
        if mang[0]>mang[-1]:
            return mang[::-1]
        return mang
    a=chuanhoa(arr1)
    b=chuanhoa(arr2)
    kq=[]
    i=0
    j=0
    while i<len(a)and j<len(b):
        if a[i]<=b[j]:
            kq.append(a[i])
            i+=1
        else:
            kq.append(b[j])
            j+=1
    kq.extend(a[i:])
    kq.extend(b[j:])
    return kq
print(merge(arr1,arr2))
```

# W7A22

```python
def timtends(dsht):
    dem_ten={}
    for i in dsht:
        ten=i.strip().split()[-1]
        dem_ten[ten]=dem_ten.get(ten,0)+1
    ten_nl=""
    slnl=0
    for ten,so_lan in dem_ten.items():
        if so_lan > slnl:
            slnl=so_lan
            ten_nl=ten
        elif so_lan==slnl:
            if ten<ten_nl:
                ten_nl=ten
    return ten_nl
n=int(input("Nhập số lượng họ và tên muốn nhập:"))
dsht=[]
for i in range(n):
    ho_ten=input()
    dsht.append(ho_ten)
print(timtends(dsht))
```
