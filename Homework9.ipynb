```python
def generate_binary(n, m, current_string=""):
    if n == 0 and m == 0:
        print(current_string)
        return
    if n > 0:
        generate_binary(n - 1, m, current_string + '0')
    if m > 0:
        generate_binary(n, m - 1, current_string + '1')
n,m=map(int,input().split())
print(generate_binary(n,m))
```

# W9A17

```python
def dtnn(arr):
    n=len(arr)
    dp=[1]*n 
    trace=[-1]*n 
    for i in range(1,n):
        for j in range(i):
            if arr[i]>arr[j] and dp[i]<dp[j]+1:
                dp[i]=dp[j]+1
                trace[i]=j
    max_len=max(dp)
    max_idx=dp.index(max_len)
    kq=[]
    while max_idx!=-1:
        kq.append(arr[max_idx])
        max_idx=trace[max_idx]
    return kq[::-1]
arr=list(map(int,input().split()))
print( dtnn(arr))
```
