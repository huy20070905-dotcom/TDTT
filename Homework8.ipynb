# W8A16

```python
class Shape:
    def area(self):
        pass
class Rectangle(Shape):
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14*self.r*self.r
loai_hinh=input().strip()
tham_so=input().strip().split()
if loai_hinh=="RECTANGLE":
    w=float(tham_so[0])
    h=float(tham_so[1])
    hinh_chu_nhat=Rectangle(w,h)
    print(f"{hinh_chu_nhat.area():.2f}")
elif loai_hinh=="CIRCLE":
    r=float(tham_so[0])
    hinh_tron=Circle(r)
    print(f"{hinh_tron.area():.2f}")
```

# W8A17

```python
class Vehicle:
    def __init__(self,brand,speed):
        self.brand=brand
        self.speed=speed
    def info(self):
        return f"{self.brand}-{self.speed}"
class Car(Vehicle):
    def __init__(self,brand,speed,seats):
        super().__init__(brand,speed)
        self.seats=seats
    def info(self):
        return f"{self.brand}-{self.speed}-{self.seats}"
loai_doi_tuong=input().strip()
du_lieu=input().strip().split()
if loai_doi_tuong=="Vehicle":
    obj=Vehicle(brand=du_lieu[0],speed=du_lieu[1])
    print(obj.info())
elif loai_doi_tuong=="Car":
    obj=Car(brand=du_lieu[0],speed=du_lieu[1],seats=du_lieu[2])
    print(obj,info())
```

```python
class Polygon:
    def perimeter(self):
        pass
class Triangle(Polygon):
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def perimeter(self):
        return self.a +self.b +self.c
a,b,c=map(int,input().split())
tam_giac=Triangle(a,b,c)
print(tam_giac.perimeter())
```

```python
class Account:
    def __init__(self,balance):
        self.balance=balance
    def deposit(self,x):
        self.balance+=x
    def get_balance(self):
        return self.balance
class SavingAccount(Account):
    def __init__(self,balance,rate):
        super().__init__(balance)
        self.rate=rate
    def add_interest(self):
        self.balance+=self.balance*self.rate
ip=input().strip().split()
if len(ip)==2:
    balance=float(ip[0])
    rate=float(ip[1])
    my_account=SavingAccount(balance,rate)
    my_account.add_interest()
    so_du_moi=my_account.get_balance()
    print(f"{so_du_moi:.2f}")
```

```python
class Matrix2x2:
    def __init__(self,a,b,c,d):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
    def trace(self):
        return self.a+self.d
    def determinant(self):
        return (self.a*self.d)-(self.b*self.c)
a,b,c,d=map(int,input().split())
ma_tran=Matrix2x2(a,b,c,d)
tr=ma_tran.trace()
det=ma_tran.determinant()
print(f"{tr} {det}")
```

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def final_price(self):
        return self.price
class DiscountProduct(Product):
    def __init__(self,name,price,percent):
        super().__init__(name,price)
        self.percent=percent
    def final_price(self):
        return self.price*(100-self.percent)/100
chuoi_nhap=input().strip().split()
if len(chuoi_nhap)==3:
    name=chuoi_nhap[0]
    price=float(chuoi_nhap[1])
    percent=float(chuoi_nhap[2])
    san_pham=DiscountProduct(name,price,percent)
    ket_qua=san_pham.final_price()
    print(f"{ket_qua:.2f}")
    
```

# W8A28

```python
class Counter:
    def __init__(self):
        self.count=0
    def inc(self):
        self.count+=1
    def dec(self):
        self.count-=1
    def value(self):
        return self.count
n=int(input().strip())
danh_sach_thao_tac=input().strip().split()
if len(danh_sach_thao_tac)==n:
    bo_dem=Counter()
    for lenh in danh_sach_thao_tac:
        if lenh=="INC":
            bo_dem.inc()
        elif lenh=="DEC":
            bo_dem.dec()
    print(bo_dem.value())
```

```python
class Time:
    def __init__(self,h,m,s):
        self.h=h
        self.m=m
        self.s=s 
    def to_seconds(self):
        return self.h*3600+self.m*60+self.s
du_lieu=input().strip().split()
if len(du_lieu)==3:
    h=int(du_lieu[0])
    m=int(du_lieu[1])
    s=int(du_lieu[2])
    gia_tri=Time(h,m,s)
    ket_qua=gia_tri.to_seconds()
    print(ket_qua)
```

# W8A30

```python
class Stack():
    def __init__(self):
        self.items=[]
    def push(self,x):
        self.items.append(x)
    def pop(self):
        self.items.pop()
    def top(self):
        return self.items[-1]
n=int(input().strip())
my_stacks=Stack()
for i in range(n):
    lenh=input().strip().split()
    if lenh[0]=="PUSH":
        giatri=int(lenh[1])
        my_stacks.push(giatri)
    elif lenh[0]=="POP":
        my_stacks.pop()
    elif lenh[0]=="TOP":
        print(my_stacks.top())
```
