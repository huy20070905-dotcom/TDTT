W2A3

```python
def g(a,b):
    if b == 0:
        print('lỗi không thể chia cho không:')
        return
    tong = a + b
    hieu = a - b
    tich = a*b
    chianguyen = a//b
    chiadu = a%b
    chia = a/b
    
    print(tong)
    print(hieu)
    print(tich)
    print(chianguyen)
    print(chiadu)
    print(f"{chia:.2f}")
a = int(input())
b = int(input())    
g(a,b)   
```
